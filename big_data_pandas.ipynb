1-Data Integration

```python
import pandas as pd
```

```python
listofcountries = ["FR2.csv","IT1.csv","IT2.csv","TH1.csv","TH2.csv","US1.csv","US2.csv"]
```

```python
houses = pd.read_csv("FR1.csv")
for i in listofcountries:
   temp = pd.read_csv(i)
   houses = houses.append(temp)
houses
```

```python
houses = houses.drop(['title','city','floor','description','bathroom_count'],axis=1)
```

```python
categorical = ['property_type','country','parking','furnished','currency']
numerical = ['square_meters','room_count','rent']
```

```python
houses.duplicated().sum()
```

```python
houses = houses.drop_duplicates()
```

```python
for i in numerical:
   x = houses[i].quantile(0.75) * 1.5
   houses = houses[houses[i]<=x]
```

4- Data Discretization 

```python
houses.describe()
```

```python
disc = sorted(houses['rent'])
```

```python
bins = [100,460,650,1085,13000,25500]
labels = ['very low','low', 'moderate', 'high', 'very high']
houses['rent_category'] = pd.cut(houses['rent'], bins=bins, labels=labels)
houses
```

5- Data transformation

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn import preprocessing
houses = pd.get_dummies(houses, columns= categorical)
scaler = MinMaxScaler()
houses[numerical] = scaler.fit_transform(houses[numerical])
le = preprocessing.LabelEncoder()
houses['rent_category'] = le.fit_transform(houses['rent_category'])
houses
```

Model Building

```python
x = houses.drop(['rent','rent_category'],axis = 1)
y = houses['rent']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
regr = RandomForestRegressor()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_train)
r2_score(y_train, y_pred)
```

```python
y_pred = regr.predict(X_test)
r2_score(y_test, y_pred)
```

No Over fitting detected

```python
regr_par = RandomForestRegressor(n_estimators=20, bootstrap = True , max_depth=10)
regr_par.fit(X_train, y_train)
y_pred = regr_par.predict(X_train)
r2_score(y_train, y_pred)
```

```python
y_pred = regr_par.predict(X_test)
r2_score(y_test, y_pred)
```
